```python
import sys
from deepforest import main
import os
from pathlib import Path
from deepforest import main
from deepforest import get_data
from deepforest.visualize import plot_results
```

```python
model = main.deepforest()
model.use_release()
```

```python
baseDir = 'trees'
content = os.listdir(baseDir)
```

```python

images = []
for file in content:
    if os.path.isfile(os.path.join(baseDir, file)) and (file.endswith('.jpg') or file.endswith('.JPG')):
        images.append(file)
```

```python
for file in images:    
    image_path = os.path.join(baseDir, file)
    rects  = model.predict_image(path=image_path, return_plot=False)
    rects.to_csv(baseDir + '/' + Path(image_path).stem + ".csv", sep=',', index=False, encoding='utf-8')
```

```python
model = main.deepforest()

model.load_model(model_name="weecology/deepforest-tree", revision="main")

sample_image_path = "trees/V2_3.jpg"
img = model.predict_image(path=sample_image_path)
plot_results(img)
```
